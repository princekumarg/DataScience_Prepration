```python
import csv
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
with open('E:/Prepration/Day1/language.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    language_counter=Counter()
    for row in csv_reader:
        language_counter.update(row['language'].split(';'))
languages=[]
popularity=[]
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
plt.bar(languages,popularity)
plt.title('languages')
plt.xlabel('programming language')
plt.ylabel('no of people')
plt.show()
```

```python
import csv
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
with open('E:/Prepration/Day1/language.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    language_counter=Counter()
    for row in csv_reader:
        language_counter.update(row['language'].split(';'))
languages=[]
popularity=[]
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
plt.barh(languages,popularity)#to convert the bar into horizontal
plt.title('languages')
plt.xlabel('programming language')
plt.ylabel('no of people')
plt.show()
```

```python
import csv
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
with open('E:/Prepration/Day1/language.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    language_counter=Counter()
    for row in csv_reader:
        language_counter.update(row['language'].split(';'))
languages=[]
popularity=[]
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
languages.reverse()
popularity.reverse()
plt.barh(languages,popularity)
plt.title('languages')
plt.xlabel('programming language')
plt.ylabel('no of people')
plt.show()
```

# Fill curve

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8,9]
y1=[2,3,7,8,9,1,4,6,7]
y2=[8,6,5,4,2,1,7,8,9]
y3=[4,7,1,2,3,4,7,3,2]
plt.plot(x,y1,'k--',label='age')
plt.plot(x,y2,'b',label='salary')
plt.plot(x,y3,'r--',label='population')
labels=['sem1','sem2','sem3']
colors=['red','yellow','green']
plt.fill_between(x,y1)
plt.legend(loc=(0.07,0.05))
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8,9]
y1=[2,3,7,8,9,1,4,6,7]
y2=[8,9,5,4,5,8,9,8,9]
y3=[4,7,1,2,3,4,7,3,2]
plt.plot(x,y1,'k--',label='age')
plt.plot(x,y2,'b',label='salary')
plt.plot(x,y3,'r--',label='population')
plt.fill_between(x,y1,alpha=0.30)
plt.fill_between(x,y2,alpha=0.30)
plt.fill_between(x,y3,alpha=0.30)
```
